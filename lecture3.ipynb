```python
a=input().split(",")
b=set(a)
c=input().split(",")
d=set(c)

e=list(b-d)
e.sort()

f=list(d-b)
f.sort()

g=list(b&d)
g.sort()

h=list(b|d)
h.sort()

print(",".join(e))
print(",".join(f))
print(",".join(g))
print(len(h))
```
