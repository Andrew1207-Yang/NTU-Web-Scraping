```python
dataList=[]
doc1.make_links_absolute(response1.url)
for eachLv1Doc in doc1.items("#side-menu > li:nth-child(n+2) > a"):
    #print("https://webscraper.io"+eachLv1Doc.attr("href"))
    print(eachLv1Doc.attr("href"))
    responseLv1=requests.get(eachLv1Doc.attr("href"))
    docLv1=pq(responseLv1.text)
    print(docLv1("#side-menu > li.active > ul > li > a").text())
    docLv1.make_links_absolute(responseLv1.url)
    for eachLv2Doc in docLv1("#side-menu > li.active > ul > li > a").items():
        responseLv2=requests.get(eachLv2Doc.attr("href"))
        docLv2=pq(responseLv2.text)
        print(eachLv2Doc.attr("href"))
        for eachItemDoc in docLv2(".thumbnail").items():
            dataDict={}
            dataDict["title"]=str(eachItemDoc(".caption>h4>.title").text())
            dataDict["price"]=float(eachItemDoc(".caption>h4.price").text().replace("$",""))
            #dataDict["price"]=float(eachItemDoc(".caption>h4.price").text()[1:])
            dataDict["stars"]=int(eachItemDoc(".ratings>p:nth-child(2)").attr("data-rating"))
            dataList.append(dataDict)
#             print(eachItemDoc(".caption>h4>.title").text())
#             print(eachItemDoc(".caption>h4.price").text())
#            #print(len(eachItemDoc(".ratings .glyphicon-star")))
#             print(eachItemDoc(".ratings>p:nth-child(2)").attr("data-rating"))
print(dataList)
            
            
```
